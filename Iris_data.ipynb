## Step - 3:

### a. Identifying the input and output : 

```python
inputs=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
inputs.head()
```

```python
target=df[['Species']]
target.head()
```

 ### b. Identifying the ML Task :
 <b> Given SepalLength, PetalLength, SepalWidth and PetalWidth,we have to predict the Species of the Iris Flower.
    
 <b> Target variable is given - Supervised.
     
 <b> Target variable is Discrete - Classification Task.

### c. Identifying the Evaluation Metric :


<b> Since the task is classification we use Accuracy Score as evaluation metric.

## Step - 4: Splitting the dataset into Training and Testing :

```python
# split into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size=0.75, random_state=100)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
type(X_train)
```

```python
type(y_train)
```

## Step - 5: Data preparation on train data:
### Numerical Variables - Standardization

```python
# scaling the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), 
                      columns = X_train.columns)

X_train.head()
```

### Categorical - LabelEncoding

```python
species_encoder = {'Iris-setosa' : 1, 'Iris-versicolor' : 2, 'Iris-virginica' : 3 }

y_train['Species'] = y_train['Species'].apply(lambda x : species_encoder[x])

y_train.head()
```

## Step - 6: Data preparation on test data:


### Numerical Variables - Standardization

```python
X_test = pd.DataFrame(scaler.transform(X_test), 
                      columns = X_test.columns)
X_test.head()
```

### Categorical - LabelEncoding

```python
species_encoder = {'Iris-setosa' : 1, 'Iris-versicolor' : 2, 'Iris-virginica' : 3 }

y_test['Species'] = y_test['Species'].apply(lambda x : species_encoder[x])

y_test.head()
```

## Step - 7: Model Training Phase 

##  KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
KNN_accuracy=accuracy_score(y_test, y_test_pred)
KNN_accuracy
```

## Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
Logistic_accuracy=accuracy_score(y_test, y_test_pred)
Logistic_accuracy
```

## Support Vector Machines

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
SVM_accuracy=accuracy_score(y_test, y_test_pred)
SVM_accuracy
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
DT_accuracy=accuracy_score(y_test, y_test_pred)
DT_accuracy
```

## Random Forest
